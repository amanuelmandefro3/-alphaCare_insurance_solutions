Univariate  Analysis

```python
#Histograms for numerical columns
numerical_cols = df.select_dtypes(include=['int64', 'float64']).columns 
df[numerical_cols].hist(figsize=(15, 10))
plt.show()
```

```python
# Bar charts for categorical columns
categorical_cols = df.select_dtypes(include=['object', 'bool']).columns
for col in categorical_cols:
    df[col].value_counts().plot(kind='bar')
    plt.title(col)
    plt.show()
```

Bivariate or Multivariate Analysis

```python
#convert TransactionMonyh to datetime
df['TransactionMonth'] = pd.to_datetime(df['TransactionMonth'])
monthly_data = df.groupby(['PostalCode', 'TransactionMonth']).agg({
    'TotalPremium': 'sum',
    'TotalClaims': 'sum'
    }).reset_index()

#plot scatter plots
plt.figure(figsize=(15, 10))
sns.scatterplot(x='TotalPremium', y='TotalClaims', hue='PostalCode', data=monthly_data)
plt.title('Scatter plot of TotalPremium vs TotalClaims')
plt.show()
```

```python
# correlation matrix

# calculate the correlation matrix
correlation_matrix =monthly_data[['TotalPremium', 'TotalClaims']].corr()

sns.heatmap(correlation_matrix, annot=True)
plt.title('Correlation matrix of TotalPremium and TotalClaims')
plt.show()
```

Data comparison

```python
# compare change in insurance cover type, permium, auto make
sns.boxplot(x='CoverType', y='TotalPremium', data=df)
plt.show()
```

Outlier Detection

```python
# Box plots fpr numerical data
for col in numerical_cols:
    df.boxplot(column=col)
    plt.title(col)
    plt.show()
```

Visualization

```python
plt.figure(figsize=(10, 6))
sns.lineplot(x='TransactionMonth', y='TotalPremium', data=df)
plt.title('Monthly Total Premium')
plt.show()

plt.figure(figsize=(10, 6))
sns.barplot(x='VehicleType', y='TotalClaims', data=df)
plt.title('Total Claims by Vehicle Type')
plt.show()

plt.figure(figsize=(10, 6))
sns.scatterplot(x='PostalCode', y='TotalPremium', hue='CoverType', data=df)
plt.title('Total Premium by Postal Code and Cover Type')
plt.show()
```
